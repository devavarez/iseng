```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
df
```

```python
df.info()
```

```python
con_y = df['Year'].isnull()
df[con_y]
```

```python
df_new = df[~con_y]
df_new
```

```python
df_copy = df_new.copy()
df_copy['Year'] = pd.to_datetime(df_copy['Year'], format='%Y', errors='coerce')
df_copy['Year'] = df_copy['Year'].dt.year
df_new = df_copy
df_new
```

```python
df_new = df_new[(df_new['Year']>= 2010) & (df_new['Year']<=2015)]
df_new
```

```python
con_null = df_new['Publisher'].isnull()
con_un = df_new['Publisher'] == 'Unknown'
df_new[con_null & con_un]
```

```python
df_new = df_new[~con_null & ~con_un]
df_new
```

```python
df_fin = df_new.groupby(['Publisher', 'Year'])[['Global_Sales', 'NA_Sales']].sum().reset_index()
df_fin
```

```python
year_sum = df_fin.groupby('Year')['Global_Sales', 'NA_Sales'].sum().reset_index()
year_sum
```

```python
con_test = df_fin['Publisher'] == 'Take-Two Interactive' #buat nyari publisher mana yg datanya ada dari 2010-2015 buat perbandingan
df_fin[con_test]
```

```python
select_pubs = df_fin.nlargest(10, 'Global_Sales')
df_select = df_fin[df_fin['Publisher'].isin(select_pubs['Publisher'])].reset_index()
df_select
```

```python
df_fin['Publisher_groups']= df_fin['Publisher'].apply(lambda x:x if x in select_pubs else 'Others')
df_fin
```

```python
plt.bar(year_sum['Year'], year_sum['Global_Sales'], alpha= 0.5, color= 'grey', label= 'Others')
plt.bar(df_select['Year'], df_select['Global_Sales'], alpha= 0.5)

color_maps = plt.colormaps['tab10']

for i, pubs in enumerate(df_select['Publisher'].unique()):
    subset = df_select[df_select['Publisher'] == pubs]
    plt.bar(subset['Year'], subset['Global_Sales'], color= color_maps(i), label= f'Publisher: {pubs}', linewidth=200)
    

plt.xlabel('Year', labelpad= 10, size= 14)
plt.ylabel('Global Sales(in million)', labelpad= 10, size= 14)
plt.title('Penjualan Video Game dari tahun 2010 hingga 2015')
plt.legend(loc='right', bbox_to_anchor=(1.6, 0.5))
plt.show()
```
